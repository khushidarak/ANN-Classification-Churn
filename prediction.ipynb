```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#load the one hot, label,scalar
model = load_model('model.h5')

#load encoder and scalar
onehot_encoder_geo = pickle.load(open('onehot_encoder_geo.pkl', 'rb'))
label_encoder_gender = pickle.load(open('label_encoder_gender.pkl', 'rb'))
scaler = pickle.load(open('scaler.pkl', 'rb'))
```

```python
#example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]])
```

```python
geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=onehot_encoder_geo.get_feature_names_out(['Geography'])
)
```

```python
input_data['Gender'] = label_encoder_gender.transform([input_data['Gender']])[0]
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df = input_df.drop('Geography', axis=1)
```

```python
input_df = pd.concat(
    [input_df, geo_encoded_df],
    axis=1
)
```

```python
input_scaled = scaler.transform(input_df)
```

```python
prediction = model.predict(input_scaled)
```

```python
print(prediction)
```

```python
if prediction[0][0] > 0.5:
    print("Customer is likely to churn.")
else:
    print("Customer is likely to stay.")
```
